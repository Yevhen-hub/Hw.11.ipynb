### 4. Build and Train a Bidirectional LSTM Model

Bidirectional RNNs process sequences in both forward and backward directions, allowing them to capture context from both past and future information in the sequence. This can be particularly beneficial for tasks like sentiment analysis.

```python
embedding_dim = 128
bidirectional_lstm_units = 64
batch_size = 32
epochs = 10

def build_bidirectional_lstm_model():
    model = Sequential([
        Embedding(max_features, embedding_dim),
        Bidirectional(LSTM(bidirectional_lstm_units)),
        Dense(1, activation='sigmoid')
    ])
    model.build(input_shape=(None, maxlen))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

bidirectional_lstm_model = build_bidirectional_lstm_model()
bidirectional_lstm_model.summary()

print("\nTraining Bidirectional LSTM Model...")
history_bidirectional_lstm = bidirectional_lstm_model.fit(
    x_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2,
    callbacks=[early_stopping]
)

loss_bidirectional_lstm, accuracy_bidirectional_lstm = bidirectional_lstm_model.evaluate(x_test, y_test, verbose=0)
print(f"\nBidirectional LSTM Test Loss: {loss_bidirectional_lstm:.4f}")
print(f"Bidirectional LSTM Test Accuracy: {accuracy_bidirectional_lstm:.4f}")
```

### 5. Build and Train a Deep RNN Model (Stacked LSTM)

For a 'deep' recurrent neural network, we'll stack multiple LSTM layers. This allows the model to learn more complex representations of the input sequence.

```python
embedding_dim = 128
deep_lstm_units_1 = 128
deep_lstm_units_2 = 64
batch_size = 32
epochs = 10

def build_deep_rnn_model():
    model = Sequential([
        Embedding(max_features, embedding_dim),
        LSTM(deep_lstm_units_1, return_sequences=True),
        LSTM(deep_lstm_units_2),
        Dense(1, activation='sigmoid')
    ])
    model.build(input_shape=(None, maxlen))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

deep_rnn_model = build_deep_rnn_model()
deep_rnn_model.summary()

print("\nTraining Deep RNN Model (Stacked LSTM)...")
history_deep_rnn = deep_rnn_model.fit(
    x_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2,
    callbacks=[early_stopping]
)

loss_deep_rnn, accuracy_deep_rnn = deep_rnn_model.evaluate(x_test, y_test, verbose=0)
print(f"\nDeep RNN Test Loss: {loss_deep_rnn:.4f}")
print(f"Deep RNN Test Accuracy: {accuracy_deep_rnn:.4f}")
```
